## Question 1: What is Simple Linear Regression?
**Simple Linear Regression** is a statistical method used to model the relationship between:
- One **independent variable (X)** → predictor/explanatory variable
- One **dependent variable (Y)** → response/outcome variable

It assumes a **linear relationship** between X and Y, meaning the change in Y can be explained by a straight-line function of X.

###Mathematical Equation
The model is represented as:


Y = beta_0 + beta_1 X + epsilon


Where:
- Y  = Dependent variable (predicted output)  
- X= Independent variable (input feature)  
- beta_0 = Intercept (value of Y when X = 0)  
- beta_1 = Slope (change in Y for one unit change in X)  
- epsilon = Error term (unexplained variation)  


### Example
Suppose we want to predict a student's **exam score (Y)** based on the **hours studied (X)**.

- X =  Hours studied  
- Y =  Exam score  
- Linear regression will try to fit a straight line that best predicts exam scores based on study hours.


## Use Cases
- Predicting house prices based on size.  
- Estimating salary based on years of experience.  
- Predicting sales based on advertising spend.  


The line represents the **best-fit linear relationship** between X and Y.

## Question 2: What are the Key Assumptions of Simple Linear Regression?

When applying **Simple Linear Regression (SLR)**, certain assumptions must be satisfied to ensure that the model gives reliable and valid results.  

### Key Assumptions

### 1. Linearity
- The relationship between the independent variable (X) and dependent variable (Y) is **linear**.  
- This means the change in Y is proportional to the change in X.  
- Example: Sales increase linearly with advertising spend.


### 2. Independence of Errors
- The residuals (errors) are independent of each other.  
- No autocorrelation should exist (important in time-series data).  

### 3. Homoscedasticity
- The variance of residuals should be **constant** across all levels of X.  
- If variance changes (fan-shape pattern), it is called **heteroscedasticity**.  


### 4. Normality of Errors
- The residuals (differences between observed and predicted values) should be **normally distributed**.  
- Important for hypothesis testing and confidence intervals.  

### 5. No Significant Outliers
- Outliers can heavily influence the regression line and distort results.  
- They should be detected and handled before fitting the model.  

###Summary
- Linearity → X and Y must have a straight-line relationship.  
- Independence → Errors must be independent.  
- Homoscedasticity → Constant variance of errors.  
- Normality → Errors follow normal distribution.  
- Outliers → Should not dominate the regression model.  

Meeting these assumptions ensures that the regression results (coefficients, predictions, significance tests) are valid and trustworthy.

## Question 3: What is Heteroscedasticity, and Why is it Important to Address in Regression Models?

### Definition
**Heteroscedasticity** occurs in a regression model when the **variance of the residuals (errors) is not constant** across all levels of the independent variable(s).  

- Ideally, regression assumes **homoscedasticity** → constant variance of errors.  
- If residuals spread out more as X increases (or decreases), we have **heteroscedasticity**.  

### Example
predicting **house prices (Y)** based on **size (X)**:
- Small houses → prediction errors are small (less variation).  
- Large houses → prediction errors are large (more variation).  
- This unequal spread of errors is **heteroscedasticity**.  

## Problems with Heteroscedasticity
1. **Violates regression assumptions** → OLS (Ordinary Least Squares) assumes constant variance.  
2. **Biased standard errors** → Leads to incorrect confidence intervals and hypothesis tests.  
3. **Unreliable significance tests (t-tests, F-tests)** → May show predictors as significant when they are not.  
4. **Reduced predictive power** → Model may underperform on new data.  

## Detecting Heteroscedasticity
- **Residual plots** → Plot residuals vs predicted values. If spread increases → heteroscedasticity exists.  
- **Statistical Tests** → Breusch-Pagan test, White’s test.  

## Fixing Heteroscedasticity
1. **Data Transformation**
   - Apply log, square root, or Box-Cox transformation to stabilize variance.  
   - Example: Use `log(Y)` instead of `Y`.  

2. **Weighted Least Squares (WLS)**
   - Give lower weights to observations with higher variance.  

3. **Robust Standard Errors**
   - Use heteroscedasticity-consistent errors (HCSE) to correct inference.

## Question 4: What is Multiple Linear Regression?

**Multiple Linear Regression (MLR)** is an extension of **Simple Linear Regression** where we model the relationship between:
- **One dependent variable (Y)**  
- **Two or more independent variables (X1, X2, ..., Xn)**  

It helps in understanding how multiple factors simultaneously influence the outcome.

###Equation


Y = beta_0 + beta_1 X_1 + beta_2 X_2 +...........+ beta_n X_n + epsilon


Where:
- Y = Dependent variable (predicted output)  
- X_1, X_2,....., X_n= Independent variables (predictors/features)  
- beta_0 = Intercept (value of Y when all Xs = 0)  
- beta_1, \beta_2, ...... , \beta_n = Coefficients (effect of each independent variable on Y)  
- epsilon = Error term (unexplained variation)  

###Example
predicting  a **car’s selling price (Y)** based on:
- **Year of manufacture (X1)**  
- **Kilometers driven (X2)**  
- **Ex-showroom price (X3)**  

The model will look like:

SellingPrice = beta_0 + beta_1(Year) + beta_2(KmDriven) + beta_3(ExShowroomPrice) + epsilon

### Advantages
- Considers multiple factors at once (more realistic than simple linear regression).  
- Provides better predictive power when multiple variables affect the outcome.  


### Limitations
- Sensitive to **multicollinearity** (when predictors are highly correlated).  
- Requires more data to estimate coefficients reliably.  
- Assumes linearity, homoscedasticity, independence, and normality of residuals.

## Question 5: What is Polynomial Regression, and how does it differ from Linear Regression?
**Polynomial Regression** is a type of regression technique that models the relationship between the **independent variable (X)** and the **dependent variable (Y)** as an **nth-degree polynomial**.  
It is used when the data shows a **non-linear relationship** but can still be approximated by a polynomial function.



### Equation


Y = beta_0 + beta_1X + beta_2X^2 + beta_3X^3 +..........+ beta_nX^n + epsilon


Where:
- Y = Dependent variable  
- X  = Independent variable  
- beta_0,beta_1,dots, beta_n = Coefficients  
- n  = Degree of the polynomial  
- epsilon = Error term  

###Example
If we want to predict the **selling price of a car** based on its **age**, a straight line (linear regression) may not fit well because prices **decrease non-linearly** with age.  
Polynomial regression can fit a **curve** to capture the actual trend.


###Difference Between Linear & Polynomial Regression

| Feature                  | Linear Regression                                | Polynomial Regression                              |
|---------------------------|------------------------------------------------|--------------------------------------------------|
| **Model Shape**          | Straight line                                   | Curved line (depends on polynomial degree)       |
| **When to Use**          | When data shows a linear relationship          | When data shows a non-linear relationship        |
| **Complexity**           | Simple and easy to interpret                   | More complex; risk of overfitting if degree is high |

### Use Cases
- Growth curves (population, bacteria growth, etc.)  
- Relationship between **car speed and stopping distance**  
- Predicting **housing prices** when the effect of area is non-linear  


### Advantages
- Captures **non-linear relationships** effectively.  
- More flexible than simple linear regression.  

### Limitations
- **Overfitting** when the polynomial degree is too high.  
- Harder to interpret compared to linear regression.  
- Sensitive to outliers.

```python
# Question 8: Implement polynomial regression on the following data:
# ● X = [1, 2, 3, 4, 5]
# 3
# ● Y = [2.2, 4.8, 7.5, 11.2, 14.7]
# Fit a 2nd-degree polynomial and plot the resulting curve.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Data
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Reshape X
X_reshaped = X.reshape(-1, 1)

# Create polynomial features (degree 2)
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X_reshaped)

# Fit the polynomial regression model
poly_model = LinearRegression()
poly_model.fit(X_poly, Y)

# Generate points for plotting the curve
X_plot = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
X_plot_poly = poly_features.transform(X_plot)
Y_plot = poly_model.predict(X_plot_poly)

# Plot the data points and the regression curve
plt.scatter(X, Y, color='red', label='Data Points')
plt.plot(X_plot.flatten(), Y_plot.flatten(), color='blue', label='Polynomial Regression Curve')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression (Degree 2)')
plt.legend()
plt.grid(True)
plt.show()
```

## Question 10: Addressing Heteroscedasticity and Multicollinearity in Real Estate Price Prediction

As a data scientist predicting house prices using features like area, number of rooms, and location, encountering heteroscedasticity and multicollinearity is common. Here are the steps I would take to address these issues and ensure a robust model:

### Addressing Heteroscedasticity

Heteroscedasticity means the variance of the residuals is not constant across all levels of the independent variables. This can lead to biased standard errors and unreliable hypothesis tests.

1.  **Detecting Heteroscedasticity:**
    *   **Residual Plots:** Plot residuals against predicted values or individual independent variables. A fan shape or cone shape in the plot indicates heteroscedasticity.
    *   **Statistical Tests:** Use formal tests like the Breusch-Pagan test or White's test to statistically confirm the presence of heteroscedasticity.

2.  **Addressing Heteroscedasticity:**
    *   **Data Transformation:** Apply transformations to the dependent variable (house price) or independent variables. Common transformations include:
        *   **Log Transformation:** Using `log(Price)` can often stabilize variance, especially if the spread of residuals increases with price.
        *   **Square Root Transformation:** Another option for stabilizing variance.
        *   **Box-Cox Transformation:** A more general transformation that can be helpful.
    *   **Weighted Least Squares (WLS):** If the pattern of heteroscedasticity is known or can be estimated, WLS can be used. This method assigns lower weights to observations with higher variance.
    *   **Robust Standard Errors:** Use heteroscedasticity-consistent standard errors (HCSE) during model estimation. This doesn't change the coefficient estimates but corrects the standard errors, leading to valid confidence intervals and hypothesis tests. Libraries like `statsmodels` in Python support this.

### Addressing Multicollinearity

Multicollinearity occurs when two or more independent variables in a regression model are highly correlated. This can make it difficult to interpret the individual coefficients and can lead to unstable model estimates. In real estate, 'Area' and 'Number of Rooms' can often be highly correlated.

1.  **Detecting Multicollinearity:**
    *   **Correlation Matrix:** Examine the correlation matrix of the independent variables. High correlation coefficients (e.g., above 0.7 or 0.8) indicate potential multicollinearity.
    *   **Variance Inflation Factor (VIF):** Calculate the VIF for each independent variable. A high VIF (often a threshold of 5 or 10 is used) suggests that the variable is highly correlated with other predictors.

2.  **Addressing Multicollinearity:**
    *   **Remove Highly Correlated Predictors:** If two variables are highly correlated, consider removing one of them from the model. Choose the variable that is less theoretically relevant or has a slightly weaker correlation with the dependent variable.
    *   **Combine Highly Correlated Predictors:** Create a new variable that is a combination of the highly correlated predictors (e.g., an index or a ratio).
    *   **Principal Component Analysis (PCA):** Use PCA to transform the correlated independent variables into a set of uncorrelated principal components. These components can then be used as predictors in the regression model.
    *   **Ridge Regression or Lasso Regression:** These are regularization techniques that can handle multicollinearity by adding a penalty term to the regression equation, which shrinks the coefficients and reduces their variance.

### Ensuring a Robust Model

After addressing heteroscedasticity and multicollinearity, I would take these steps to ensure a robust model:

*   **Re-evaluate Model Assumptions:** After making adjustments, re-check the assumptions of linearity, independence, and normality of residuals using residual plots and statistical tests.
*   **Model Evaluation:** Evaluate the model's performance using appropriate metrics (e.g., R-squared, Adjusted R-squared, Mean Squared Error, Root Mean Squared Error) on a separate test set.
*   **Cross-Validation:** Use cross-validation techniques to assess the model's generalization ability and ensure it performs well on unseen data.
*   **Domain Expertise:** Consult with real estate experts to ensure the model's findings are consistent with industry knowledge and make sense in the context of house pricing.
*   **Iterative Process:** Model building is an iterative process. I would continue to refine the model based on evaluation results and feedback.

By systematically addressing heteroscedasticity and multicollinearity, along with following best practices for model evaluation and validation, I can build a more reliable and robust regression model for predicting house prices.
