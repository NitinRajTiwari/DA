## Question 1: Explain the differences between AI, ML, Deep Learning (DL), and Data Science (DS).
### **1. Artificial Intelligence (AI)**
- **Definition:** AI is the broad field of computer science focused on building systems that can perform tasks that typically require human intelligence.
- **Scope:** It includes reasoning, problem-solving, natural language understanding, vision, robotics, and decision-making.
- **Example:** Virtual assistants like Siri or Alexa that understand and respond to voice commands.


### **2. Machine Learning (ML)**
- **Definition:** ML is a subset of AI that focuses on developing algorithms that allow computers to **learn from data** and improve performance without being explicitly programmed.
- **Scope:** Deals mainly with supervised learning, unsupervised learning, and reinforcement learning.
- **Example:** Spam email detection, where the system learns to classify emails based on patterns in the data.

### **3. Deep Learning (DL)**
- **Definition:** DL is a specialized subset of ML that uses **neural networks with multiple layers (deep architectures)** to model complex relationships in data.
- **Scope:** Particularly powerful in handling large amounts of unstructured data like images, audio, and text.
- **Example:** Self-driving cars use DL for object detection and scene understanding.


### **4. Data Science (DS)**
- **Definition:** DS is a multidisciplinary field that combines **statistics, programming, data analysis, and domain knowledge** to extract meaningful insights and knowledge from data.
- **Scope:** Focuses on the complete data pipeline – collection, cleaning, exploration, visualization, modeling, and communication of results.
- **Example:** Analyzing customer purchasing behavior to recommend products on e-commerce platforms.

##  Question 2: What are the types of Machine Learning? Describe each with one real-world example.

Machine Learning (ML) is broadly categorized into **three main types**, and sometimes a **fourth category** (Reinforcement Learning) is considered separately.  

## 1. Supervised Learning  
- **Definition**: In supervised learning, the model is trained on a labeled dataset, meaning the input data (features) and the output (target/label) are both known.  
- **Goal**: Learn the mapping function from input `X` to output `Y`.  
- **Techniques**:  
  - Classification (predict categories)  
  - Regression (predict continuous values)  

**Example**:  
- Predicting house prices based on features like size, location, and number of bedrooms (Regression).  
- Email spam detection (Classification).  


## 2. Unsupervised Learning  
- **Definition**: In unsupervised learning, the model works on an unlabeled dataset, where only inputs `X` are given without any output `Y`.  
- **Goal**: Find hidden patterns, structure, or groupings in the data.  
- **Techniques**:  
  - Clustering  
  - Dimensionality reduction  

**Example**:  
- Customer segmentation in marketing (grouping customers with similar buying behavior).  
- Market basket analysis (finding products often bought together).  


## 3. Semi-Supervised Learning  
- **Definition**: Uses both labeled and unlabeled data. Typically, a small portion of data is labeled, and a large portion is unlabeled.  
- **Goal**: Improve learning accuracy when labeling is expensive or time-consuming.  

**Example**:  
- Medical diagnosis systems where only a few medical records are labeled by doctors, but the system learns patterns using a mix of labeled and unlabeled records.  

## 4. Reinforcement Learning (RL)  
- **Definition**: An agent learns by interacting with an environment, receiving feedback in the form of rewards or penalties.  
- **Goal**: Maximize cumulative rewards by taking the best actions in an environment.  

**Example**:  
- Self-driving cars learning to navigate roads.  
- AlphaGo (Google’s AI) beating human players in the game of Go.  


| **Type**              | **Data Used** | **Goal** | **Real-World Example** |
|------------------------|--------------|----------|-------------------------|
| **Supervised**         | Labeled      | Predict outcomes | House price prediction |
| **Unsupervised**       | Unlabeled    | Find hidden patterns | Customer segmentation |
| **Semi-Supervised**    | Mix (few labeled + many unlabeled) | Improve learning | Medical diagnosis |
| **Reinforcement**      | Rewards/Penalties | Learn by interacting | Self-driving cars |

## Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.

### **1. Overfitting**
- **Definition**: Overfitting happens when a machine learning model learns the training data *too well*, including noise and random fluctuations.  
- The model performs **very well on training data** but **poorly on unseen (test) data**.  
- It usually occurs when the model is **too complex** (e.g., too many parameters or too many decision tree splits).

**Example**:  
A decision tree that memorizes all training samples, including outliers, but fails to generalize to new data.  


### **2. Underfitting**
- **Definition**: Underfitting occurs when a machine learning model is **too simple** and fails to capture the underlying patterns in the data.  
- The model performs **poorly on both training and test data**.  
- It usually happens when the model is **not complex enough** or the training is insufficient.  

**Example**:  
A linear regression line trying to fit a curved relationship (non-linear data).  

### **3. Bias-Variance Tradeoff**
- **Bias**: Error due to **wrong assumptions** or oversimplified models.  
  - High bias → underfitting.  
- **Variance**: Error due to **model sensitivity to small fluctuations** in training data.  
  - High variance → overfitting.  

The tradeoff:
- **High Bias (Simple Model)** → Low variance, High bias → Poor accuracy.  
- **High Variance (Complex Model)** → Low bias, High variance → Poor generalization.  
- **Goal**: Find a balance between bias and variance that minimizes the total error.

## Question 4: What are outliers in a dataset, and list three common techniques for handling them?

### Outliers
- **Definition**:  
  Outliers are data points that differ significantly from the majority of other observations in a dataset.  
  They can occur due to variability in the data, measurement errors, or unusual events.  

- **Example**:  
  In a dataset of students' exam scores (out of 100), most scores range between 50–90.  
  If one student scores **5** or **100**, these values might be considered **outliers**.


### Why Outliers are Important?
- They can **distort statistical measures** like mean and standard deviation.
- They may **negatively impact machine learning models**, especially those sensitive to scale (e.g., Linear Regression, KNN).
- In some cases, they may represent **important rare events** (e.g., fraud detection, disease outbreaks).


### Common Techniques for Handling Outliers

1. **Removal of Outliers**  
   - Remove extreme values that are unlikely to be valid data points.  
   - Example: Dropping data points beyond **3 standard deviations from the mean**.  

2. **Transformation / Winsorization**  
   - Apply mathematical transformations (e.g., **log, square root**) to reduce the effect of outliers.  
   - Winsorization replaces extreme values with a **threshold value** (e.g., capping the top 1% to the 99th percentile).  

3. **Imputation / Binning**  
   - Replace outliers with **median** or **mean** values of the feature.  
   - Grouping values into **bins** (e.g., age groups) can also reduce outlier impact.  

- **Outliers** = unusual data points far from the majority of data.  
- **Handling techniques** = Removal, Transformation, or Imputation.  
- Choice of method depends on the **problem context** and whether outliers represent **errors** or **important signals**.

## Question 5: Explain the process of handling missing values and mention one imputation technique for numerical and one for categorical data.

### Handling Missing Values

In real-world datasets, missing values are common and need to be addressed before applying machine learning models. If not handled properly, missing data can lead to biased results or errors in model training.

### Process of Handling Missing Values:
1. **Identify Missing Values**  
   - Use functions like `isnull()` or `info()` in pandas to check missing data.  

2. **Analyze the Impact**  
   - Check the proportion of missing values in each column.  
   - If a column has too many missing values (e.g., >50%), consider dropping it.  

3. **Decide on Strategy**  
   - **Drop rows/columns** → If missing values are few and not critical.  
   - **Imputation** → Replace missing values with meaningful estimates.  


### Imputation Techniques

### 1. **For Numerical Data**
   - **Mean/Median Imputation** → Replace missing values with the column mean or median.
   - Example: If age values are missing, replace them with the average age of the dataset.

   ```python
   df['Age'].fillna(df['Age'].mean(), inplace=True)
   ```

### 2. For Categorical Data

  - Mode Imputation → Replace missing values with the most frequent (mode) value.

  - Example: If "Gender" is missing, fill with the most frequent category ("Male" or "Female").
```python
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
```

````python
# Question 8: Write a Python script to:
# ● Generate 1000 samples from a normal distribution.
# ● Introduce 50 random missing values.
# ● Fill missing values with the column mean.
# ● Plot a histogram before and after imputation.
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# 1000 samples from a noraml dist. with mean 50 and SD 10
data=np.random.normal(loc=50,scale=10,size=1000)
````

````python
nan_indices=np.random.choice(range(0,1001),size=50,replace=True)
nan_indices
````

````python
data_with_nan=data.copy()
data_with_nan[nan_indices]=np.nan
````

````python
mean_value=np.nanmean(data_with_nan)
data_imputed=np.where(np.isnan(data_with_nan),mean_value,data_with_nan)
````

````python
#Plot histogram before and after imputation
plt.figure(figsize=(12, 5))

# Histogram before imputation
plt.subplot(1, 2, 1)
sns.histplot(data_with_nan[~np.isnan(data_with_nan)], bins=30, color="skyblue", edgecolor="black",kde=True)
plt.title("Histogram Before Imputation")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Histogram after imputation
plt.subplot(1, 2, 2)
sns.histplot(data_imputed, bins=30, color="lightgreen", edgecolor="black",kde=True)
plt.title("Histogram After Imputation")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
````

## Question 10: Data Preparation Plan for Retail Customer Transaction Dataset

As a data scientist working with a **customer transaction dataset** containing missing values, outliers, imbalance, and categorical variables, the goal is to prepare the data properly before training a machine learning model.

**step-by-step plan**:

---

## 1. Handling Missing Data
- **Problem:** Some customers have missing `Age` values.
- **Solution:**
  - For **numerical data** like `Age` → Use **Mean/Median imputation** (Median preferred if skewed).
  - Alternatively, use **KNN Imputation** if dataset is large and we want more accuracy.
- **Example:** If 20% of Age values are missing, replace them with the median age of customers.

---

## 2. Handling Outliers
- **Problem:** Transaction Amount contains extreme values (e.g., very large amounts).
- **Techniques:**
  - **IQR Method (Interquartile Range):** Detect values below Q1 - 1.5*IQR or above Q3 + 1.5*IQR and treat them as outliers.
  - **Z-score Method:** Remove/Cap values beyond ±3 standard deviations.
  - **Transformation:** Apply **log/square root transformation** if data is skewed.
- **Decision:** Instead of removing, apply **capping (winsorization)** so important transactions are not lost.

---

## 3. Handling Imbalanced Target (Fraud vs. Non-Fraud)
- **Problem:** Fraud cases are much fewer than non-fraud.
- **Techniques:**
  - **Resampling:**
    - **Oversampling:** Use **SMOTE (Synthetic Minority Oversampling Technique)** to create synthetic fraud samples.
    - **Undersampling:** Randomly reduce non-fraud cases.
  - **Algorithmic Approaches:**
    - Use **class weights** in models (e.g., Logistic Regression, Random Forest, XGBoost).
- **Decision:** Use **SMOTE + Class Weights** to balance dataset.

---

## 4. Encoding Categorical Variables
- **Problem:** Features like `Payment Method` are categorical.
- **Techniques:**
  - **One-Hot Encoding:** For nominal variables (e.g., `Credit Card`, `Debit Card`, `UPI`).
  - **Label Encoding:** For ordinal features (if categories have order, e.g., `Low`, `Medium`, `High`).
- **Decision:** Apply **One-Hot Encoding** for `Payment Method` since it’s nominal.

---

## 5. Final Data Preparation Pipeline
1. **Handle Missing Values** → Median imputation for `Age`.
2. **Treat Outliers** → Capping extreme `Transaction Amount` values.
3. **Balance Data** → Apply SMOTE + Class Weights.
4. **Encode Categorical Features** → One-Hot Encoding for `Payment Method`.
5. **Feature Scaling** → Standardize/Normalize numerical features like `Age`, `Transaction Amount`.
6. **Train-Test Split** → Split into 70-80% training and 20-30% testing set.
