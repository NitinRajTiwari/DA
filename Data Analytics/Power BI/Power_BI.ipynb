### Q1.Define Power BI and What are the key components of the Power BI ecosystem? Briefly explain:
 ● Power BI Desktop

 ● Power BI Service

 ● Power BI Mobile

 ● Power BI Gateway


## Power BI

**Power BI** is a **business intelligence (BI) and data visualization tool** developed by Microsoft.  
It helps users **collect, transform, analyze, and visualize data** to create interactive dashboards and reports for better business decision-making.

### **1. Power BI Desktop**
- A **free desktop application** used to:
  - Connect to multiple data sources
  - Clean and transform data (using Power Query)
  - Build data models and visual reports
- Used mainly by **data analysts and report creators**

Example: Creating sales dashboards from Excel or SQL data

### **2. Power BI Service**
- A **cloud-based platform** (PowerBI.com)
- Used to:
  - Publish and share reports
  - Create dashboards
  - Schedule data refresh
  - Collaborate with teams

Example: Sharing live dashboards with management

### **3. Power BI Mobile**
- Mobile apps available for **Android, iOS, and Windows**
- Allows users to:
  - View dashboards and reports on the go
  - Get real-time alerts and notifications
Example: Sales managers checking KPIs on mobile phones


### **4. Power BI Gateway**
- A bridge between **on-premises data sources** and **Power BI Service**
- Enables:
  - Secure data refresh
  - Live queries to local databases (SQL Server, Oracle)
Example: Accessing on-premise company data from Power BI Service


Power BI makes data **easy to understand, interactive, and actionable**.

## Question 2: Comparison of Power BI Visuals

Power BI offers different types of visuals for effective data representation. Choosing the right visual depends on the **type of data** and **analysis goal**.

### Pie Chart vs Donut Chart

### **Pie Chart**
- Shows **parts of a whole** as slices.
- Best when there are **few categories**.

**When to use:**
To display percentage contribution clearly  
When exact proportions are important

**Example:**
- Market share of companies (Apple, Samsung, Others)


### **Donut Chart**
- Similar to a Pie Chart but with a **hole in the center**.
- Allows extra information (like total value) in the center.

**When to use:**
When you want a modern look  
When you want to display totals or labels inside

**Example:**
- Expense breakdown with total expenses shown in the center


### **Preference**
- Use **Pie Chart** for simple proportion comparison
- Use **Donut Chart** when you want **additional context** or better aesthetics


##  **Bar Chart vs Column Chart**

### **Bar Chart**
- Bars are **horizontal**
- Works well with **long category names**

**When to use:**
 Comparing values across many categories  
 When labels are long and readability matters

**Example:**
- Sales by product category (long product names)


### **Column Chart**
- Bars are **vertical**
- Ideal for **time-based or trend comparison**

**When to use:**
 Showing changes over time  
 Comparing values across few categories

**Example:**
- Monthly sales comparison

## Question 3: Data Modeling Concepts in Power BI

Good data modeling is essential in Power BI for **accurate analysis, better performance, and easy maintenance**.

## **Star Schema vs Snowflake Schema**

### **Star Schema**
- Consists of:
  - One **central fact table**
  - Multiple **dimension tables** connected directly to it
- Simple structure and easy to understand
- Faster query performance

**When to use:**
Recommended for Power BI models  
Suitable for reporting and dashboards

**Example:**
- FactSales connected to DimCustomer, DimProduct, DimDate


### **Snowflake Schema**
- Dimension tables are **normalized** into multiple related tables
- More complex relationships
- Slower performance compared to star schema

**When to use:**
When storage optimization is required  
When dimensions are very large and complex

**Example:**
- DimProduct → DimCategory → DimSubCategory

## **Primary Key vs Foreign Key (Power BI Relationships)**

### **Primary Key**
- Uniquely identifies each record in a table
- Contains **unique and non-null values**

**Example:**
- CustomerID in Customer table


### **Foreign Key**
- References a Primary Key in another table
- Used to establish relationships

**Example:**
- CustomerID in Sales table


## Why Cardinality Is Important

**Cardinality** defines the relationship between tables:
- One-to-One (1:1)
- One-to-Many (1:*)
- Many-to-One (*:1)
- Many-to-Many (*:*)

### **Importance of Cardinality**
Ensures correct data aggregation  
Prevents duplication of values  
Affects filter direction and DAX calculations  
Improves report accuracy

**Example:**
- One customer → Many sales (1:*)

## Question 4: Differentiate

## **1️ Calculated Column vs Measure**

### **Calculated Column**
- Calculated **row by row**
- Stored physically in the table
- Computed during **data refresh**
- Uses **row context** by default
- Increases model size

**Example (Calculated Column):**
```DAX
TotalPrice = Sales[Quantity] * Sales[UnitPrice]
```
Use when:
You need values at row level
Used in slicers, filters, or relationships

###**Measure**
- Calculated on the fly

- Not stored in the table

- Computed during report interaction

- Uses filter context

- More memory efficient

**Example (Measure):**

```DAX
Total Sales = SUM(Sales[TotalPrice])
```
Use when:
You need aggregated values
Used in visuals (charts, cards, tables)

Calculated Column vs Measure (Comparison)
| Feature           | Calculated Column | Measure         |
| ----------------- | ----------------- | --------------- |
| Calculation Level | Row-level         | Aggregate-level |
| Storage           | Stored            | Not stored      |
| Evaluation Time   | Data refresh      | Query time      |
| Context Used      | Row Context       | Filter Context  |


## **2 Row Context and Filter Context**
###Row Context
- Refers to the current row being evaluated

- Automatically applies in calculated columns

**Example:**

```DAX
DiscountedPrice = Sales[Price] * 0.9
```
Each row gets its own discounted price

###Filter Context
- Applies filters based on:

- Visuals

- Slicers

- Page/report filters

- Used mainly by measures

**Example:**

```DAX
Total Sales = SUM(Sales[Price])
```
If you apply a slicer for Year = 2023,
-the measure recalculates sales only for 2023

## Question 5: Difference Between a Report and a Dashboard in Power BI

In Power BI, **Reports** and **Dashboards** serve different purposes in data visualization and decision-making.

## Power BI Report

### **What is a Report?**
- A **report** is a detailed, multi-page collection of visuals.
- Created using **Power BI Desktop**.
- Based on a **single dataset**.

### **Key Features**
- Can have **multiple pages**
- Interactive (filters, slicers, drill-down)
- Rich and detailed analysis

### **Example**
- A sales report with pages for:
  - Monthly sales
  - Region-wise performance
  - Product-wise analysis


## Power BI Dashboard

### **What is a Dashboard?**
- A **dashboard** is a **single-page** visual summary.
- Created in **Power BI Service**.
- Can combine visuals from **multiple reports/datasets**.

### **Key Features**
- High-level overview
- Less interactive than reports
- Ideal for executives and quick insights

### **Example**
- CEO dashboard showing:
  - Total revenue
  - Top products
  - Sales trends


## Report vs Dashboard (Comparison)

| Feature | Report | Dashboard |
|-------|--------|-----------|
| Pages | Multiple | Single |
| Created In | Power BI Desktop | Power BI Service |
| Data Source | One dataset | Multiple datasets |
| Interaction | High | Limited |
| Purpose | Detailed analysis | Quick overview |
