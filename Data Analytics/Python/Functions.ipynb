#Q1 What is the difference between a function and a method in Python?
In Python, both **functions** and **methods** are used to perform a task, but they are different in terms of how they are defined and called.

- **Function:** A block of reusable code defined using `def`, which is independent and can be called directly.
- **Method:** A function that is associated with an object and is called using dot (`.`) notation.

**Example:**
```python
#Function

def greet():
    print("Hello!")
greet()  # Calling function


#Method (associated with string object)

s = "hello"
print(s.upper())  # 'upper' is a method
```

#Q5. What are iterators in Python and how do they differ from iterables?

- An **iterable** is any Python object that can be looped over (like `list`, `tuple`, `str`).
- An **iterator** is an object with a `__next__()` method that returns the next item.

You can get an iterator from an iterable using the `iter()` function.

**Example:**

```python
nums = [1, 2, 3]
it = iter(nums)

print(next(it))  # Output: 1
print(next(it))  # Output: 2
```
So, iterables can be converted into iterators, and iterators maintain state between calls to next().

# Q6. Explain the concept of generators in Python and how they are defined.

A **generator** is a special type of iterator in Python that is defined using the `yield` keyword instead of `return`. It generates values lazily — one at a time — and does not store the entire sequence in memory.

**Defining a generator:**
```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1
```
**Using the generator:**
```python
for num in countdown(3):
    print(num)
# Output: 3, 2, 1
```
Generators are memory-efficient and suitable for large datasets or infinite sequences.

#Q 7. What are the advantages of using generators over regular functions?

```markdown
Generators offer several advantages:

1. **Memory efficiency:** They yield items one at a time, avoiding large memory usage.
2. **Lazy evaluation:** They compute values on demand.
3. **Better performance:** Especially useful in loops or data streams.

**Example Comparison:**

Regular function:
```python
def get_numbers():
    return [i for i in range(1000000)]
  
def get_numbers():
    for i in range(1000000):
        yield i
```
The generator does not build a full list in memory, saving resources.

#Q8. What is a lambda function in Python and when is it typically used?

A **lambda function** is an anonymous function defined using the `lambda` keyword. It can have any number of arguments but only one expression.

Typically used for short, throwaway functions, especially with `map()`, `filter()`, and `sorted()`.

**Example:**
```python
# Lambda to add two numbers
add = lambda x, y: x + y
print(add(3, 4))  # Output: 7

# With sorted()
names = ['John', 'Alex', 'Chris']
sorted_names = sorted(names, key=lambda name: len(name))
```

#Q9. Explain the purpose and usage of the `map()` function in Python.

The `map()` function applies a given function to all items in an iterable (like a list) and returns a `map` object (which is an iterator).

**Syntax:**
```python
map(function, iterable)
nums = [1, 2, 3, 4]
squared = map(lambda x: x**2, nums)
print(list(squared))  # Output: [1, 4, 9, 16]
```
map() is useful for element-wise operations on collections.

#Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

All three are functional programming tools that operate on iterables:

1. **map():** Applies a function to every item.
```python
map(lambda x: x*2, [1, 2, 3])  # [2, 4, 6]
```

2. **filter():** Filters items based on a condition.
```python
filter(lambda x: x%2 == 0, [1, 2, 3, 4])  # [2, 4]
```
3. **reduce():** Reduces a list to a single value using a function (imported from functools).
```python
from functools import reduce
reduce(lambda x, y: x + y, [1, 2, 3, 4])  # 10
```

- map() - transforms

- filter() - selects

- reduce() - aggregates

````python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
#    terms.
class FibonacciIterator:
    def __init__(self, n):
        self.max_terms = n  # Maximum terms to generate
        self.count = 0      # Current position
        self.a = 0                  # First number
        self.b = 1                  # Second number

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration  # End

        # For the first two numbers, return a and b
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            # for rest numbers
            next_value = self.a + self.b
            self.a, self.b = self.b, next_value
            self.count += 1
            return next_value

fib=FibonacciIterator(10)
for num in fib:
  print(num,end=" ")
````

````python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def power_of_two(n):
  for i in range(1,n+1):
    yield 2**i

p=power_of_two(10)
for i in p:
  print(i,end=", ")
# we can also do
# print(next(p))    and so on
````

````python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()

# Suppose 'sample.txt' is a file in the same directory
for line in read_file_lines("sample.txt"):
    print(line)

# sample.txt                  #output
# Hello                       Hello
# World                       World
# This is a file              This is a file
````
