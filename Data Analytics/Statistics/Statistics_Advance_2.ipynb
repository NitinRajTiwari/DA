## Question 1: What is Hypothesis Testing in Statistics?

**Hypothesis Testing** is a statistical method used to make decisions or draw conclusions about a population parameter based on sample data.  

It helps us test whether the observed data provides enough evidence to reject a **null hypothesis (H₀)** in favor of an **alternative hypothesis (H₁)**.



###Key Concepts:
1. **Null Hypothesis (H₀):**
   - The default assumption.
   - Example: "There is no difference in average sales before and after a new campaign."

2. **Alternative Hypothesis (H₁):**
   - What we want to prove.
   - Example: "The average sales increased after the new campaign."

3. **Test Statistic:**
   - A standardized value (like Z, T, Chi-square, F) calculated from the sample data.

4. **P-value:**
   - Probability of observing the sample data if the null hypothesis is true.
   - If `p-value < α (significance level)`, we reject H₀.

5. **Significance Level (α):**
   - The threshold probability to decide rejection of H₀.
   - Commonly used values: 0.05 (5%), 0.01 (1%).

### Steps in Hypothesis Testing:
1. State the null (H₀) and alternative (H₁) hypotheses.  
2. Choose significance level (α).  
3. Select the test (Z-test, T-test, etc.) and compute the test statistic.  
4. Calculate the **p-value**.  
5. Make a decision:  
   - If `p-value < α` → Reject H₀ (evidence supports H₁).  
   - If `p-value ≥ α` → Fail to reject H₀ (not enough evidence).  

### Example:
- H₀: The average customer satisfaction score = 7.  
- H₁: The average customer satisfaction score ≠ 7.  
- We collect survey data, calculate the test statistic and p-value, then make a decision.

## Question 2: What is the Null Hypothesis, and how does it differ from the Alternative Hypothesis?

In **hypothesis testing**, we always begin with two competing statements:

1. **Null Hypothesis (H₀):**
   - It is the default assumption that **there is no effect, no difference, or no relationship** in the population.  
   - It represents the "status quo" or baseline condition.  
   - Example: *"The new marketing strategy has no effect on sales."*

2. **Alternative Hypothesis (H₁ or Ha):**
   - It is the claim that **contradicts the null hypothesis**.  
   - It states that there **is an effect, a difference, or a relationship**.  
   - Example: *"The new marketing strategy increases sales."*

### Key Differences:

| Aspect | Null Hypothesis (H₀) | Alternative Hypothesis (H₁) |
|--------|----------------------|------------------------------|
| Meaning | No effect / no difference | There is an effect / difference |
| Role | Default assumption | Competing claim |
| Example | μ = μ₀ (population mean equals hypothesized mean) | μ ≠ μ₀, μ > μ₀, or μ < μ₀ |
| Testing Goal | Attempt to reject H₀ | Provide evidence in favor of H₁ |

---

**In summary:**  
- We never "prove" H₀; we only test whether the data provides **enough evidence to reject H₀** in favor of H₁.

## Question 3: Explain the Significance Level in Hypothesis Testing and its Role in Deciding the Outcome of a Test

In hypothesis testing, the **Significance Level (α)** is the threshold we set to decide whether to reject the null hypothesis (H₀).  

### Definition:
- The significance level (α) is the **probability of rejecting the null hypothesis when it is actually true** (Type I Error).  
- It represents the **tolerance for risk of a false positive conclusion**.  

### Common Values of α:
- **0.05 (5%)** → Most commonly used in practice.  
- **0.01 (1%)** → Stricter, used in highly sensitive studies (e.g., medical trials).  
- **0.10 (10%)** → Less strict, used in exploratory research.  

### Role in Hypothesis Testing:
1. Set α (e.g., 0.05).  
2. Conduct the test and compute the **p-value**.  
3. Compare p-value with α:  
   - If **p-value ≤ α** → Reject H₀ (evidence supports H₁).  
   - If **p-value > α** → Fail to reject H₀ (insufficient evidence against H₀).  

---

### Example:
- Suppose α = 0.05.  
- If the p-value from a t-test = 0.03 → Since 0.03 < 0.05, we **reject H₀**.  
- This means the result is **statistically significant**.  

The significance level (α) controls the risk of making a **Type I Error**, and it is the benchmark for deciding whether the observed data provides enough evidence to reject the null hypothesis.

## Question 4: What are Type I and Type II Errors? Give Examples of Each

In hypothesis testing, errors can occur because we are making decisions based on sample data (which may not perfectly represent the population). The two main types of errors are:

### 1. **Type I Error (False Positive)**
- **Definition**: Rejecting the null hypothesis (H₀) when it is actually true.  
- **Probability**: Equal to the significance level **α**.  
- **Analogy**: "Convicting an innocent person."  

**Example:**  
- Null Hypothesis (H₀): A medicine has no effect on curing a disease.  
- Alternate Hypothesis (H₁): The medicine cures the disease.  
- If we conclude the medicine works when it actually doesn’t, we make a **Type I error**.  

### 2. **Type II Error (False Negative)**
- **Definition**: Failing to reject the null hypothesis (H₀) when it is actually false.  
- **Probability**: Denoted by **β**.  
- **Analogy**: "Letting a guilty person go free."  

**Example:**  
- Null Hypothesis (H₀): A machine is producing items with correct weight.  
- Alternate Hypothesis (H₁): The machine is producing items with incorrect weight.  
- If the machine is faulty but we conclude it is fine, we make a **Type II error**.  

| Error Type   | Decision Made                      | Reality (True State)        | Analogy                        |
|--------------|-------------------------------------|-----------------------------|--------------------------------|
| **Type I**   | Reject H₀ (say effect exists)      | H₀ is actually true         | Convicting an innocent person  |
| **Type II**  | Fail to reject H₀ (say no effect)  | H₀ is actually false        | Freeing a guilty person        |

- **Type I Error (α)**: False positive (rejecting true H₀).  
- **Type II Error (β)**: False negative (failing to reject false H₀).

# Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

### **Z-test**
- A **Z-test** is a statistical test used to determine whether there is a significant difference between sample and population means when the **population variance is known**.
- It is based on the **standard normal distribution (Z-distribution)**.
- Typically used for **large sample sizes (n > 30)**.
  
**Formula:**
Z = {bar{X} -mu}/{sigma /sqrt{n}}

Where:  
- bar{X} = sample mean  
- mu = population mean  
- sigma = population standard deviation  
- n = sample size  

### **T-test**
- A **T-test** is used to determine if there is a significant difference between the sample mean and the population mean (or between two sample means) when the **population variance is unknown**.
- It uses the **Student’s t-distribution**.
- Typically used for **small sample sizes (n ≤ 30)**.  
- Variance is estimated from the sample itself.

**Formula:**

t = { bar{X} - mu}/{s /sqrt{n}}


Where:  
- bar{X} = sample mean  
- mu = population mean  
- s = sample standard deviation  
- n = sample size  

---

### **Key Differences**

| Feature              | **Z-test**                             | **T-test**                         |
|----------------------|-----------------------------------------|-------------------------------------|
| **Distribution**     | Standard Normal (Z-distribution)       | Student’s t-distribution            |
| **Population Variance** | Known                              | Unknown (estimated from sample)     |
| **Sample Size**      | Large samples (\(n > 30\))             | Small samples (\(n ≤ 30\))          |
| **Use Case**         | Comparing means when population variance is known | Comparing means when variance is unknown |

---

### **Example Scenarios**
- **Z-test Example:** Testing whether the average height of students in a large school matches the national average, given that the population variance is known.  
- **T-test Example:** Testing whether a new teaching method improves scores in a class of 20 students, without knowing the population variance.

```python
# Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
# Python. Show the Python code and interpret the results.
# sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
# 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
# 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
# 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Convert to numpy array
data = np.array(sample_data)

# Known population mean
mu = 50

# Sample mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # unbiased std (denominator n-1)
n = len(data)

# Compute Z-statistic
z_stat = (sample_mean - mu) / (sample_std / np.sqrt(n))

# p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Standard Deviation: {sample_std:.4f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject Null Hypothesis: The mean is significantly different from 50.")
else:
    print("Fail to Reject Null Hypothesis: No significant difference from 50.")
```

```python
# Question 8: Write a Python script to simulate data from a normal distribution and
# calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=50, scale=5, size=1000)  # mean=50, std=5, n=1000

# Step 2: Calculate sample mean and standard error
sample_mean = np.mean(data)
std_error = stats.sem(data)

# Step 3: Compute 95% confidence interval
confidence = 0.95
confidence_interval = stats.norm.interval(confidence, loc=sample_mean, scale=std_error)

# Step 4: Print results
print("Sample Mean:", sample_mean)
print("95% Confidence Interval:", confidence_interval)

# Step 5: Plot histogram of the data
plt.figure(figsize=(8,5))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', density=True, alpha=0.7)

# Overlay normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, sample_mean, np.std(data))
plt.plot(x, p, 'r', linewidth=2)

# Add vertical lines for confidence interval
plt.axvline(confidence_interval[0], color='green', linestyle='--', label="Lower CI")
plt.axvline(confidence_interval[1], color='green', linestyle='--', label="Upper CI")
plt.axvline(sample_mean, color='blue', linestyle='-', label="Sample Mean")

plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
# Question 9: Write a Python function to calculate the Z-scores from a dataset and
# visualize the standardized data using a histogram. Explain what the Z-scores represent
# in terms of standard deviations from the mean.
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = [(x - mean) / std for x in data]
    return z_scores

# normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=1000)  # mean=50, std=5, size=1000

# Z-scores
z_scores = calculate_z_scores(data)

# histogram of Z-scores
plt.hist(z_scores, bins=30, color='skyblue', edgecolor='black', density=True)
plt.title("Histogram of Z-Scores (Standardized Data)")
plt.xlabel("Z-Score")
plt.ylabel("Frequency Density")
plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
plt.legend()
plt.show()
```

### Explanation:

- **Z-score formula**:  

  Z = {X - mu}/{sigma}

  where:
  - X = individual data point  
  - mu = mean of dataset  
  - sigma = standard deviation  

- **Interpretation**:  
  - A **Z-score of 0** means the data point is exactly at the mean.  
  - A **positive Z-score** means the data point is above the mean.  
  - A **negative Z-score** means the data point is below the mean.  
  - For example, a **Z-score of +2** means the value is **2 standard deviations above the mean**.  

- Standardizing data (converting to Z-scores) makes it easier to compare datasets with different scales, and it is widely used in hypothesis testing, outlier detection, and machine learning.
