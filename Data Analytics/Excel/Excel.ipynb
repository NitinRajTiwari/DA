## Question 1: Difference between **Paste** and **Paste Special** in Excel

### **Paste**
- **Paste** simply inserts the copied data exactly as it is.
- It includes **values, formulas, formatting, and comments** together.

**Example:**
If you copy a cell containing the formula `=A1+B1` and paste it using **Paste**, the formula will also be pasted and recalculated in the new cell.

### **Paste Special**
- **Paste Special** gives you **control** over what you want to paste.
- You can paste **only values, only formulas, only formatting**, or perform operations like transpose.

**Common Paste Special options:**
- **Values** → Pastes only the final result (no formula)
- **Formulas** → Pastes only the formula
- **Formats** → Pastes only formatting (color, font, borders)
- **Transpose** → Converts rows into columns and vice versa

**Example:**
If you copy a cell with the formula `=A1+B1` and use **Paste Special → Values**, only the calculated number (e.g., `50`) will be pasted, not the formula.

### **Summary**
- **Paste** → Copies everything as-is.
- **Paste Special** → Allows selective pasting based on your need.

This makes **Paste Special** very useful for data cleaning, reporting, and formatting in Excel.

## Question 2: Freeze Panes vs Split Panes in Excel

### **Freeze Panes**
- **Freeze Panes** is used to keep specific rows or columns visible while scrolling through a worksheet.
- Commonly used when working with **large datasets**.

**Usefulness:**
- Helps you keep **headers (row titles or column names)** visible at all times.
- Makes data comparison easier while scrolling.

**Example:**
If the first row contains column headings (Name, Sales, Date), using  
**View → Freeze Panes → Freeze Top Row**  
will keep the header row visible when you scroll down.



### **Split Panes**
- **Split Panes** divides the worksheet into **multiple independent sections**.
- Each section can be scrolled separately.

**Usefulness:**
- Useful for **comparing data from different parts** of the same worksheet.
- Helps view distant rows/columns side-by-side.

**Example:**
Using **View → Split**, you can scroll the top part to see summary data and the bottom part to view detailed records at the same time.


### **Summary**
- **Freeze Panes** → Locks rows/columns so they stay visible.
- **Split Panes** → Splits the screen into multiple scrollable areas.

Both features improve navigation and readability when handling large Excel datasets.

## Question 3: Difference Between Inserting a New Row and a New Column in Excel

### **Inserting a New Row**
- A **new row** is added **horizontally**.
- Existing rows **shift downward** to make space.
- Used when you want to add a **new record or entry**.

**Example:**
If you insert a row above Row 5, the existing Row 5 and below will move down.


### **Inserting a New Column**
- A **new column** is added **vertically**.
- Existing columns **shift to the right**.
- Used when you want to add a **new field or attribute**.

**Example:**
If you insert a column before Column C, the existing Column C and after will move right.



### **Inserting Multiple Rows or Columns at Once**
Yes, Excel allows inserting **multiple rows or columns simultaneously**.

**How to do it:**
- Select the same number of rows or columns you want to insert.
- Right-click → **Insert**.

**Example:**
Selecting 3 rows and clicking Insert will add 3 new rows at once.
Similarly, selecting 2 columns will insert 2 new columns.


### **Summary**
- **Rows** → Horizontal data insertion.
- **Columns** → Vertical data insertion.
- **Multiple rows/columns** can be inserted in one action, saving time.

## Question 4: Logical Functions in Excel

### **What are Logical Functions?**
Logical functions in Excel are used to **make decisions** by testing conditions and returning different results based on whether the condition is **TRUE or FALSE**.  
They help automate decision-making and analysis in spreadsheets.


### **Common Logical Functions and Their Applications**

### **1. IF()**
- Tests a condition and returns one value if TRUE and another if FALSE.

**Syntax:**
```excel
=IF(condition, value_if_true, value_if_false)

=IF(A1 >= 40, "Pass", "Fail")
```
### **2. AND()**

- Returns TRUE only if all conditions are TRUE.

Example:
``` excel
=AND(A1 >= 40, B1 >= 40)
```
Used when multiple criteria must be satisfied.

### **3. OR()**

Returns TRUE if any one condition is TRUE.

Example:
```excel
=OR(A1 >= 40, B1 >= 40)

```
Used when at least one condition is sufficient.

### **4. NOT()**

Reverses the logical value.

Example:
```excel
=NOT(A1 > 50)
```

Used to negate conditions.

###**5. Nested IF**

Multiple IF statements inside one another.

Example:
```excel
=IF(A1 >= 75, "Distinction", IF(A1 >= 50, "Pass", "Fail"))

```
Used for multi-level decision-making.

## Question 5: Purpose of XLOOKUP and Difference from VLOOKUP

### **Purpose of XLOOKUP**
**XLOOKUP** is a modern Excel lookup function used to **search for a value in a range** and return a corresponding result from another range.  
It is designed to **replace older lookup functions** like VLOOKUP and HLOOKUP with more flexibility and fewer limitations.

---

### **XLOOKUP Syntax**
```excel
=XLOOKUP(lookup_value, lookup_array, return_array)
```
### Difference

| Feature           | XLOOKUP                  | VLOOKUP                  |
| ----------------- | ------------------------ | ------------------------ |
| Lookup Direction  | Can search left or right | Only searches right      |
| Column Index      | Not required             | Required (column number) |
| Exact Match       | Default                  | Needs FALSE              |
| Error Handling    | Built-in (if_not_found)  | Requires IFERROR         |
| Table Structure   | Flexible                 | Fixed structure          |
| Row/Column Lookup | Supports both            | Vertical only            |

Example

Using VLOOKUP:
```excel
=VLOOKUP(A2, A2:D10, 3, FALSE)
```

Using XLOOKUP:
```excel
=XLOOKUP(A2, A2:A10, C2:C10)
```


### XLOOKUP is Better because

    1.No need to count column numbers

    2.Works even if columns are rearranged

    3.Supports exact match by default

    4.Can return custom message if value not found

## Question 10: Detecting and Handling Missing Data in Excel

When working as a **Data Scientist**, handling missing data is an important data cleaning step. Excel provides several built-in tools and functions to **identify and manage missing values** effectively.

---

## Detecting Missing Values in Excel

### **1. Go To Special**
- Path: **Home → Find & Select → Go To Special → Blanks**
- This highlights all empty cells in the selected range.

**Use case:**
- Quickly locate missing values in large datasets  
- Useful for bulk operations like filling or deleting blanks

### **2. ISBLANK() Function**
- Checks whether a specific cell is empty.

**Syntax:**
```excel
=ISBLANK(A1)
```
Output:

TRUE → Cell is empty

FALSE → Cell has data

Use case:
- Logical checks
- Conditional formatting
- Creating flags for missing values

### 3. COUNTBLANK() Function
Counts the number of blank cells in a range.

Syntax:

```excel
=COUNTBLANK(A1:A100)
```
Use case:
- Understand the extent of missing data
- Data quality assessment before analysis

### Handling Missing Values in Excel
1. Replace Missing Values
- Use Find & Replace to fill blanks with:

- Mean/Median (for numerical data)

- "Not Available" or "Unknown" (for categorical data)

2. Fill Blanks Using Formulas
```excel
=IF(ISBLANK(A1), AVERAGE(A:A), A1)
```
3. Delete Rows with Missing Data
- Select blanks using Go To Special

- Right-click → Delete → Entire Row

4. Forward Fill / Backward Fill
- Useful for time-series data

- Use Fill Down / Fill Up
